```julia
using JuMP
using Cbc

exo1b=Model(optimizer_with_attributes(Cbc.Optimizer))

@variable(exo1b,x[1:3])

@objective(exo1b,Min,2*x[1]-3*x[2]+x[3])

@constraint(exo1b,constraint1,-2*x[1]+x[2]+x[3]==2)
@constraint(exo1b,constraint2,x[1]-x[3]==3)
@constraint(exo1b,constraint3,x[1]+x[2]<=1)
@constraint(exo1b,constraint4[i in 2:3],x[i]>=0)



JuMP.optimize!(exo1b)

println(exo1b)

for i in 1:3
    println("Valeur de la variable x",i,": ", JuMP.value(x[i]))
end

println("Valeur optimale: ", objective_value(exo1b))
```
