```julia
using JuMP
using Cbc

exo1c=Model(optimizer_with_attributes(Cbc.Optimizer))

@variable(exo1c,x[1:2],Int)

@objective(exo1c,Min,2*x[1]-x[2])

@constraint(exo1c,constraint1,x[1]+3*x[2]==2)
@constraint(exo1c,constraint2,-2*x[1]+5*x[2]<=7)
@constraint(exo1c,constraint3,x[1]+x[2]<=4)
@constraint(exo1c,constraint4,x[1]>=0)

JuMP.optimize!(exo1c)

println(exo1c)

for i in 1:2
    println("Valeur de la variable x",i,": ",JuMP.value(x[i]))
end

println("Valeur optimale: ", objective_value(exo1c))
```
