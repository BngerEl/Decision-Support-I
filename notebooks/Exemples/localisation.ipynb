```julia
using JuMP
using Cbc

localisation=Model(optimizer_with_attributes(Cbc.Optimizer))

@variable(localisation,x[1:4, 1:3]>=0)
@variable(localisation,y[1:4],Bin)

@objective(localisation,Min,150*y[1] + 200*y[2] + 100*y[3] + 150*y[4] +
20*x[1,1]+ 40*x[1,2] + 60*x[1,3] + 30*x[2,1]+ 10*x[2,2] + 50*x[2,3] + 30*x[3,1]
+ 20*x[3,2]+ 10*x[3,3] + 10*x[4,1] + 10*x[4,2] + 20*x[4,3])

@constraint(localisation,capacitéS1,20*y[1]-x[1,1]-x[1,2]-x[1,3]>=0)
@constraint(localisation,capacitéS2,20*y[2]-x[2,1]-x[2,2]-x[2,3]>=0)
@constraint(localisation,capacitéS3,10*y[3]-x[3,1]-x[3,2]-x[3,3]>=0)
@constraint(localisation,capacitéS4,10*y[4]-x[4,1]-x[4,2]-x[4,3]>=0)

@constraint(localisation,demandeM1,x[1,1]+x[2,1]+x[3,1]+x[4,1]>=10)
@constraint(localisation,demandeM2,x[1,2]+x[2,2]+x[3,2]+x[4,2]>=15)
@constraint(localisation,demandeM3,sum(x[i,3] for i in 1:4)>=10)

JuMP.optimize!(localisation)

println(localisation)

for i in 1:4, j in 1:3
    println("Quantité transportée depuis S",i," vers le marché M",j,": ",JuMP.value(x[i,j]))
end
println()
for i in 1:4
    println("Usine sur le site S",i,": ",JuMP.value(y[i]))
end

println("Coût total: ",objective_value(localisation))
```
