```julia
using JuMP
using Cbc

bateaux=Model(optimizer_with_attributes(Cbc.Optimizer))

@variable(bateaux,x[1:2],Int) #bateaux


@objective(bateaux,Max,2400*x[1]+2000*x[2])

@constraint(bateaux,Toiles,4*x[1]<=400)
@constraint(bateaux,Moteurs,x[2]<=120)
@constraint(bateaux,Fibres,8*x[1]+4*x[2]<=1000)

JuMP.optimize!(bateaux)

println(bateaux)
println("Nombre de bateaux à voile: ",JuMP.value(x[1]))
println("Nombre de bateaux à moteur: ",JuMP.value(x[2]))

println("Profit maximum: ",objective_value(bateaux))
```
